## Basics Of Plotting

```python
import matplotlib.pyplot as plt
```

```python
y = [1,4,9,16,25]
x = [1,2,3,4,5]
```

```python
plt.plot(x,y)
plt.show()
```

### Customization in plots

```python
plt.plot(x, y, color='green', marker='p', linestyle='dashed', linewidth=2, markersize=12)
plt.show()
```

```python
plt.plot(x, y, 'r*-' ,lw=2, ms=25)
plt.show()
```

### Graph with custom axes

```python
plt.plot(x, y, color='green', marker='p', linestyle='dashed', linewidth=2, markersize=12)
plt.xlim([0,8])
plt.ylim([0,40])
plt.show()
```

```python
plt.plot(x, y, color='green', marker='p', linestyle='dashed', linewidth=2, markersize=12)
plt.axis([0,6,0,30])
l,lv = plt.xticks()
plt.title("Simple graph")
plt.xlabel("N -> 1,2,3")
plt.ylabel("N^2 -> 1,4,9")
plt.show()
```

```python
plt.plot(x, y, color='green', marker='p', linestyle='dashed', linewidth=2, markersize=12)
plt.title("Simple graph")
plt.xlabel("N -> 1,2,3")
plt.ylabel("N^2 -> 1,4,9")
plt.show()
```

### Subplots

```python
import numpy as np
def f(t):
    return np.exp(-t) * np.cos(2 * np.pi * t)


t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)



fig, (ax1, ax2) = plt.subplots(1, 2,figsize=(20, 10))
ax1.text(1,0.5,"test")
ax1.plot(t1, f(t1), 'bo', t2, f(t2), 'k')



ax2.plot(t2, np.cos(2 * np.pi * t2), 'r--')
plt.show()
```

### Pie Chart

```python
fruits = ["apple","apple","apple",'mango','mango','papaya']

import collections
data = collections.Counter(fruits)

plt.pie(data.values(),labels=data.keys(),labeldistance=0.5)
plt.show()
```

### Bar Chart

```python
plt.bar(data.keys(),data.values(),color='green')
plt.show()
```

```python
n = 12
X = np.arange(n)
Y1 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)
Y2 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)

plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white')
plt.bar(X, -Y2, facecolor='#ff9999', edgecolor='white')

for x, y in zip(X, Y1):
    plt.text(x + 0.4, y + 0.05, '%.2f' % y, ha='center', va='bottom')

plt.ylim(-1.25, +1.25)
plt.show()
```

## Seaborn Visualization 

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

### Histograms Or Density Plots

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.distplot(tips.total_bill)
```

```python
sns.distplot(tips.total_bill,hist=False)
```

### Scatter Plots

```python
sns.relplot(x="total_bill", y="tip", data=tips)
```

### Adding third dimension using color

```python
sns.relplot(x="total_bill", y="tip", hue="day",data=tips)
```

### Adding fourth dimension through style

```python
sns.relplot(x="total_bill", y="tip",hue="day", style="time", data=tips)
```

### Adding Fifth Dimension Size

```python
sns.relplot(x="total_bill", y="tip",hue="day", style="time",size="size", data=tips,height=5,aspect=3)
```

### Modifying Color Sheets

```python
cmap = sns.cubehelix_palette(dark=.3, light=.8, as_cmap=True)
sns.relplot(x="total_bill", y="tip", hue="day", size="smoker",palette="Set2",data=tips)
```

### Regression Plots

```python
sns.lmplot(x="total_bill", y="tip", data=tips)
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips)
```

### Line Plot

```python
# Load an example dataset with long-form data
fmri = sns.load_dataset("fmri")
fmri.head()
```

```python
sns.relplot(x="timepoint", y="signal", kind="line", data=fmri)
```

```python
# Plot the responses for different events and regions
sns.lineplot(x="timepoint", y="signal",
             hue="region", style="event",
             data=fmri)
```

### Categorical Plots

### Bar Plots

```python
import pandas
names = ['mpg','cylinders','displacement','horsepower','weight','acceleration','model_year','origin','car_name']
cars = pandas.read_csv("..\data\\auto-mpg.csv",sep="\s+",names=names)


cars['maker'] = cars.car_name.map(lambda x: x.split()[0])
cars.origin = cars.origin.map({1: 'America', 2: 'Europe', 3: 'Asia'})
cars=cars.applymap(lambda x: None if x == '?' else x).dropna()
cars['horsepower'] = cars.horsepower.astype(float)
cars.head()
```

```python
sns.catplot(data=cars, x="model_year", y="mpg",kind="bar",height=5,aspect=3)
```

```python
sns.catplot(data=cars, x="model_year", y="mpg",hue="origin",kind="bar",height=5,aspect=3)
```

### Point Plot

```python
sns.catplot(data=cars, x="model_year", y="mpg",kind="point",height=5,aspect=3)
```

```python
sns.catplot(data=cars, x="model_year", y="mpg",hue="origin",kind="point",height=5,aspect=3)
```

```python
sns.catplot(data=cars, x="model_year", y="mpg",row="origin",kind="point")
```
