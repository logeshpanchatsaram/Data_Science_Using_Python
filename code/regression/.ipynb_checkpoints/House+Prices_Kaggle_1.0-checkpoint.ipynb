#### The **SalesPrice** data seems to be skewed. The variable can be transformed using log transformation.

### Finding Missing Values

```python
msno.matrix( house_df[house_df.columns[pd.isnull(house_df).any()].tolist()] , p = 10, sort = "descending" )
```

```python
house_df_clean = house_df.drop(["Id", "Alley", "PoolQC", "MiscFeature"], axis = 1 )
```

```python
numeric_cols = house_df_clean.dtypes[house_df_clean.dtypes != "object"].index
```

```python
numeric_cols
```

```python
house_df_clean["SalePrice"] = np.log1p(house_df_clean["SalePrice"])
```

```python
sns.distplot( house_df_clean.SalePrice)
```

### Transform the variables which are highly skewed

```python
from scipy.stats import skew

skewed_cols = house_df_clean[numeric_cols].apply(lambda x: skew(x.dropna())) #compute skewness
skewed_cols = skewed_cols[skewed_cols > 0.75]
skewed_cols = skewed_cols.index

house_df_clean[skewed_cols] = np.log1p(house_df_clean[skewed_cols])
```

```python
house_df_clean.head()
```

### Create Dummy Variables for Categorical Variables

```python
category_cols = house_df_clean.dtypes[house_df_clean.dtypes == "object"].index
```

```python
category_cols
```

```python
house_all_df = pd.get_dummies( house_df_clean )
```

```python
house_all_df.columns
```

```python
house_all_df = house_all_df.fillna( house_all_df.mean() )
```

```python
len( house_all_df.columns )
```

### Spit the dataset into train and test

```python
Y = house_all_df['SalePrice']
X = house_all_df.drop( 'SalePrice', axis = 1 )
```

```python
from sklearn.cross_validation import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split( X, Y, test_size = 0.3, random_state = 42 )
```

## Build a linear regression model

```python
from sklearn.linear_model import LinearRegression


lreg = LinearRegression()
lreg.fit( X_train, Y_train )
```

### Validate the model and measure accuracy

```python
y_pred = lreg.predict( X_test )
```

### R-square and RMSE

```python
from sklearn import metrics

metrics.r2_score( Y_test, y_pred )
```

```python
np.sqrt( metrics.mean_squared_error( Y_test, y_pred ) )
```

### Draw the coefficient values

```python
def draw_coefficient( model, columns ):
    
    all_coefs = pd.Series( model.coef_, index = columns )

    coef_values_df = pd.concat([all_coefs.sort_values().head(10),
                                all_coefs.sort_values().tail(10)])

    mplt.rcParams['figure.figsize'] = (6.0, 8.0)
    coef_values_df.plot(kind = "barh")
    plt.title("Coefficients Values" )
```

```python
draw_coefficient( lreg, X_train.columns )
```

## KFold Cross validation 

```python
from sklearn import cross_validation
```

```python
lreg = LinearRegression()

lreg_cv_scores = cross_validation.cross_val_score( lreg, X_train, Y_train, scoring = "mean_squared_error", cv = 5 )
```

```python
np.mean( list( map( lambda x: np.sqrt( -x ), lreg_cv_scores ) ) )
```

## Build a Ridge Regression Model

```python
from sklearn.linear_model import Ridge, RidgeCV, ElasticNet, LassoCV
```

```python
r_model = Ridge( alpha = 10 )
r_model.fit( X_train, Y_train )
Y_pred_r = r_model.predict( X_test )
metrics.r2_score( Y_test, Y_pred_r )
```

```python
np.sqrt( metrics.mean_squared_error( Y_test, Y_pred_r ) )
```

```python
draw_coefficient( r_model, X_train.columns )
```

## Grid Search For Finding Best Alpha Parameter

```python
from sklearn.grid_search import GridSearchCV
```

```python
parameters = {'alpha':[0.01, 0.1, 1, 5, 10, 15, 20, 30]}
grid = GridSearchCV(r_model,parameters, scoring = "mean_squared_error" , cv=5)
grid.fit(X_train, Y_train)
```

```python
grid.grid_scores_
```

```python
grid.best_score_
```

```python
[scores[0]['alpha'] for scores in grid.grid_scores_]
```

```python
ridge_rmse = pd.Series( [np.sqrt( -scores[1] ) for scores in grid.grid_scores_], 
                index =  [scores[0]['alpha'] for scores in grid.grid_scores_] ) 
```

```python
ridge_rmse
```

```python
ridge_rmse.min()
```

### RMSE Vs. Alpha Parameters

```python
ridge_rmse.plot(title = "Ridge RMSE Vs. alpha")
plt.xlabel("Alpha Values")
plt.ylabel("RMSE")
```

## Build a Lasso Model

```python
from sklearn.linear_model import Lasso
```

```python
l_model = Lasso( alpha = 0.001 )
l_model.fit( X_train, Y_train )
Y_pred_l = l_model.predict( X_test )
metrics.r2_score( Y_test, Y_pred_l )
```

```python
np.expm1( Y_pred_l[0:10] )
```

```python
np.expm1( Y_test[0:10] )
```

```python
draw_coefficient( l_model, X_train.columns )
```

### Check Variable Shrinkage

```python
sum( l_model.coef_ == 0.00 )
```

```python
l_model = Lasso()
parameters = {'alpha':[0.00001, 0.00005, 0.0001, 0.001, 0.005]}
l_grid = GridSearchCV(l_model,parameters, scoring = "mean_squared_error" , cv=5)
l_grid.fit(X_train, Y_train)

l_grid.grid_scores_
```

```python
l_grid.best_score_
```

```python
l_estimate = l_grid.best_estimator_
```

```python
selected_features = X_train.columns[ l_estimate.coef_ > 0 ]
```

```python
[scores[0]['alpha'] for scores in l_grid.grid_scores_]

lasso_rmse = pd.Series( [np.sqrt( -scores[1] ) for scores in l_grid.grid_scores_], 
                index =  [scores[0]['alpha'] for scores in l_grid.grid_scores_] ) 

lasso_rmse

lasso_rmse.plot(title = "Lasso RMSE Vs. alpha")
plt.xlabel("Alpha Values")
plt.ylabel("RMSE")
```

```python
lasso_rmse.min()
```
